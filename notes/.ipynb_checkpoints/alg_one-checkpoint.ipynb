### Implements algorithm 1 in the parti-game [paper](https://link.springer.com/content/pdf/10.1007/BF00993591.pdf)

Learn a controller from a start region to a goal region on a continuous space; Four increasingly effective algorithms to partition discrete state spaces. Algorithm 1 and 2 are non-learning; 3 and 4 learn, hence explore, the world while planning a route to the goal. Here, I implement algorithm 1

```python
def cell_grads(g, data, dims=None, derivFunc=None):
    """
     [derivC, derivL, derivR] = computeGradients(g, data, derivFunc, upWind)

     Estimates the costate p at position x for cost function data on grid g by
     numerically taking partial derivatives along each grid direction.
     Numerical derivatives are taken using the levelset toolbox

     Inputs: grid      - cell structure; cell slice within the overall grid from 
                         where we must update the value function
             data      - value function slice on the concerned cell, grid
             derivFunc - derivative approximation function (from level set
                         toolbox)
     Output: derivC    - (central) gradient in a g.dim by 1 vector
             derivL    - left gradient
             derivR    - right gradient

    Lekan Molu, August 11, 2021
    """
    if not dims:
        dims = zeros(g.dim, 1)
        dims.fill(True)

    if not derivFunc:
        # this computes the gradients of the value function
        derivFunc = upwindFirstWENO5


    # Go through each dimension and compute the gradient in each dim
    derivC = cell(g.dim, 1)

    if numDims(data) == g.dim:
        tau_length = 1
    elif numDims(data) == g.dim + 1:
        tau_length = size(data)
        tau_length = tau_length[-1]
    else:
        error('Dimensions of input data and grid don''t match!')

    # Just in case there are NaN values in the data (usually from TTR functions)
    numInfty = 1e6
    data[np.isnan(data)] = numInfty

    # Just in case there are inf values
    data[np.isinf(data)] = numInfty

    for i in range(g.dim):
        if dims[i]:
            derivC[i] = zeros(size(data))

            ## data at a single time stamp
            if tau_length == 1:
                # Compute gradient using upwinding weighted essentially non-oscillatory differences
                derivL, derivR = derivFunc(g, data, i)

                # Central gradient
                derivC[i] = 0.5*(derivL + derivR)
            else:
                ## data at multiple time stamps
                for t in range(tau_length):
                    derivL, derivR = derivFunc(g, data[t, ...], i)
                    derivC[i][t,...] = 0.5*(derivL + derivR)

            # Change indices where data was nan to nan
            derivC[i][nanInds] = np.nan

            # Change indices where data was inf to inf
            derivC[i][infInds] = np.inf

    return derivC, derivL, derivR
```

```python
from Utilities import isfield

class NonLearning():
    def __init__(self, agent, grid, grid_cells):
        """
        Implements Algoirithm One of the Parti Games Paper.
        This is the so-called nonlearning algorithm.
        
        Cells  are axis-aligned hyperrectangles.  For 2D grids, the states 
        are 2D with coords (x,y). Cells are discrete entities with indices 
        given by grid_cells.idx. Their corresponding neighbors are given by
        grid_cellls.neighs.

        This algo. takes an environmental model, here a dynamical system, 
        and a partition P P, in this case grid_cells computed from cell_neighs.py.

        Attrs:
            dynSys: Dynamical System that defines the geometrical state of the agent.
            Given a state, we integrate it forward in time e.g. RK4 so as to give us the state
            at the next time step under the influence of a local controller.   

            grid: overall grid containing all the cells             

            grid_cells: grid cells containing initial partitioning of the cells.

            See cells_from_grid function
            action_space: list of cell indices for every cell in the grid g
         """
        self.agent = agent
        self.grid = grid
        self.grid_cells = grid_cells
        self.cur_cell = None
        self.cur_time = 0
        self.stuck_time = 300 # after this time, the agent is declared stuck in a cell
        self.action_space = list(range(len(grid_cells)))
        pass
    
    def reset_time():
        self.cur_time = 0
    
    def stuck(self):
        """
            Test for stuck is taken to be if we exceed time t in a 
            given cell after applying the local greedy controller
        """
        if self.cur_time > self.stuck_time:
            return True
        else: return False
    
    def find_center(self, cell):
        """
            Given a cell within a state space, calculate its 
            center coordinates. 
        """
        return np.squeeze((cell.max - cell.min)/2)
        
            
    def next(self, cell, goal_cell):
        return next_partition(self.find_center(cell), goal_cell)
        

    def next_partition(self, cur_cell, goal_cell):
        """
        Given the parameters defined in the function arguments,
        appy a local greedy controller that ains at cell goal_cell
        until we exit cur_cell or we become stuck in cell cur_cell. 
        Inputs
            cur_x: current state of the agent
            goal_cell: cell we are trying to navigate towards
        """
        goal_center = self.find_center(goal_cell)
        self.cur_cell = cur_cell
        while not self.stuck:
            # aim for goal_center
            # only compute gradient for current hyperrectangle
            Deriv = cell_grads(self.grid, goal_cell);
            deriv = eval_u(self.gridg, Deriv, self.agent.x)
            uOpt = self.agent.get_opt_u(t=0, deriv=deriv, uMode='min', y=0)
            # evolve the systen
            xt = self.agent.update_state(uOpt, dtSmall, self.agent.x, d=[])
            
            return cell
            cur_x = self.control(cur_x, goal_center)
            
            self.cur_time += 1
        
        xgoal = [goal_center[0], goal_center[1], ]

        
    def partigame_one(self, dynSys, gs):
        """
             Cells with indices given by gdict.keys() are axis-aligned hyperrectangles.
             For the 2D grid above, the states ate 2D with coords (x,y)
             Cells are discrete entities with indices given by gdict.keys() 
             Neighs(Cell_i) are defined by the neighs_2d/3d function in grids

             Alg 1 takes an environmental model, here a dynamical system, and a partition P

             Inputs:
                 dynSys: Dynamical System that defines the geometrical state of the agent.
                 Given a statem we integrate it forward in time e.g. RK4 so as to give us the state
                 at the next time step under the influence of a local controller. I use an LQR.

                 part: cell partition within the grid that we are starting from
                 This bundle object vontains all the possible neighbors of the goal_cell

             We will use a Dubins car model for ourt dynamical system
         """
        # accumulate the possible neighbors of all cells within the state
        pass
    

# partigame_one(dynSys, parti) 
```
